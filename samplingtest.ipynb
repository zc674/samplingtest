```python
import googlemaps
import pandas as pd
```

```python
df=pd.read_csv("/Users/zianchen/Downloads/sampling2test.csv",encoding = "ISO-8859-1")
```

```python
gmaps_key=googlemaps.Client(key="mykey")
```

```python
df["1stop"]=None
df["2ndop"]=None
```

```python
for i in range(0,len(df),1):
    geocode_result=gmaps_key.places_nearby((df.iat[i,6],df.iat[i,7]), rank_by="distance",type="establishment")
    try:
        foption=geocode_result["results"][0]['name']
        soption=geocode_result["results"][1]['name']
        df.iat[i,13]=foption
        df.iat[i,14]=soption
    except:
        foption=None
        soption=None
```
